````python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
````

````python
# Loading the dataset
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
````

````python
# Removing rows which have missing values
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))#len refers to Total no.of rows
````

````python
# Dividing the dataset into features (X) and the target variable (y)
X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])#we need to estimate the price of the car
````

````python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('Make')#Manufacturer
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
````

````python
# Setting up subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: Box plot of car prices based on body style
sns.boxplot(x='body-style', y='price', data=data, ax=axs[0])
axs[0].set_title('Box Plot for Car Prices and Body Style')
axs[0].set_xlabel('Body Style')
axs[0].set_ylabel('Price')

# Second subplot: Violin plot of car prices based on body style
sns.violinplot(x='body-style', y='price', data=data, ax=axs[1], inner='quartile')
axs[1].set_title('Violin Plot for Car Prices and Body Style')
axs[1].set_xlabel('Body Style')
axs[1].set_ylabel('Price')

# Adjust layout for better fit
plt.tight_layout()
plt.show()
````

````python
# Setting up subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: Box plot of car prices based on body style
sns.boxplot(x='body-style', y='price', data=data, ax=axs[0])
axs[0].set_title('Box Plot for Car Prices and Body Style')
axs[0].set_xlabel('Body Style')
axs[0].set_ylabel('Price')

# Second subplot: Swarm plot of car prices based on body style
sns.swarmplot(x='body-style', y='price', data=data, ax=axs[1])
axs[1].set_title('Swarm Plot for Car Prices and Body Style')
axs[1].set_xlabel('Body Style')
axs[1].set_ylabel('Price')

# Adjust layout for better fit
plt.tight_layout()
plt.show()
````
