````python
# numpy.
import numpy as np
from numpy import linalg
from numpy.linalg import norm
from scipy.spatial.distance import squareform, pdist

# sklearn.
import sklearn
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale

# Random state.
RS = 20150101

# matplotlib.
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import matplotlib
%matplotlib inline

# seaborn.
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('muted')
sns.set_context("notebook", font_scale=1.5,
                rc={"lines.linewidth": 2.5})
````

````python
digits = load_digits()
digits.data.shape
````

````python
nrows, ncols = 2, 5
plt.figure(figsize=(6,3))
plt.gray()
for i in range(ncols * nrows):
    ax = plt.subplot(nrows, ncols, i + 1)
    ax.matshow(digits.images[i,...])
    plt.xticks([]); plt.yticks([])
    plt.title(digits.target[i])
````

````python
# We first reorder the data points according to the handwritten numbers.
X = np.vstack([digits.data[digits.target==i]
               for i in range(10)])
y = np.hstack([digits.target[digits.target==i]
               for i in range(10)])
````

````python
digits_proj = TSNE(init="pca", random_state=RS).fit_transform(X)
````

````python
def scatter(x, colors):
    # We choose a color palette with seaborn.
    palette = np.array(sns.color_palette("hls", 10))

    # We create a scatter plot.
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40,
                    c=palette[colors.astype(int)])
    plt.xlim(-25, 25)
    plt.ylim(-25, 25)
    ax.axis('off')
    ax.axis('tight')

    # We add the labels for each digit.
    txts = []
    for i in range(10):
        # Position of each label.
        xtext, ytext = np.median(x[colors == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=24)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)
    plt.show()

    return f, ax, sc

scatter(digits_proj, y)
````

How does the computational complexity and runtime of t-SNE scale with dataset size and dimensionality?

```
# Answer
```
* Small Datasets (n < 10,000): Exact t-SNE works well but might still it is slow for very high dimention
* Medium-Sized Datasets (10,000 < n < 100,000): Barnes-Hut t-SNE is a good practical choice.
* Large Datasets (n > 100,000): FIt-SNE or other approximation techniques like UMAP might be preferred.
##Time complexities of these Algorithms is:
* Exact ---> O(n^2)
* Barnes-Hut t-SNE ---> O(nlogn)
* FIt-SNE (Fourier-Interpolated t-SNE) ---> O(n⋅d+n⋅logn)
* Exact t-SNE suffers significantly from increased dimensionality, as pairwise similarity matrices grow in complexity, affecting runtime and memory.
* Barnes-HUt and FIt-SNE are less affected by d, as they approximate or interpolate pairwise relationships. However, preprocessing the data to reduce d often benefits runtime and accuracy.
* if the dimentions are higher than we will maximum try to reduce them to a decent number we can use UMAP algoritm also as it is more efficient for high dimentional data sets


What are some limitations or potential pitfalls to be aware of when using t-SNE? (tell atleast 3)


```
Answer
```
 * Sensitivity to Hyperparameters
     * Perplexity effects the relation ship between local and global points if we give perplexity poor it can mislead embeddings
     * Pitfall : Selecting appropriate values requires experimentation, as there are no one-size-fits-all settings. and tuning can lead the parameters to not to reflect the true structure of the data.
 * Non-Deterministic Results
     * Makes it harder to distinguish between meaningful variations and randomness as the output changes with random seed.     
 * Limited Interpretability
     * t-SNE prioritizes preserving local neighborhoods, often at the expense of global relationships.
     * Embeddings may overemphasize local clustering while obscuring larger-scale data trends.
     * Misleading visualizations can lead to incorrect conclusions about the data structure.
 *  Computational Complexity
 * Crowding Problem
     * High-dimensional data often has points that are far apart in the original space but are forced into a smaller area in low-dimensional space.
     * Pitfall: The algorithm struggles to accurately represent both local and global relationships simultaneously.
     * So, it Leads to the loss of meaningful information, especially in datasets with a mix of dense and sparse clusters.
* Lack of Scalability
     * it Requires preprocessing, approximation techniques, or switching to alternatives like UMAP as it cant handle larger data sets as it is not scalable and it is computationally more expensive.
